## drop unuseful data
- some data like url, unnamed & ... is not useful, so we can remove them from the dataframe.

```python
spotify_yt_df_removed_column=spotify_yt_df.copy()
```

```python
remove_column=['Unnamed: 0', 'Url_spotify','Uri','Url_youtube','Description','Key']
spotify_yt_df_removed_column = spotify_yt_df_removed_column.drop(remove_column , axis=1 )
```

```python
spotify_yt_df_removed_column.shape
```

```python
spotify_yt_df_removed_column['official_video'].unique()
```

```python
spotify_yt_df_removed_column['Licensed'].unique()
```

#removing NAN value from columns by Dtype=object

```python
categorical_cols = spotify_yt_df_removed_column.select_dtypes('object').columns.tolist()
categorical_cols
```

```python
for i in categorical_cols:
    spotify_yt_df_removed_column_NANvalues=spotify_yt_df_removed_column.dropna(subset=[i])
```

```python
spotify_yt_df_removed_column_NANvalues.shape
```

```python
spotify_yt_df_removed_column_NANvalues.info()
```

#Imputing missing numeric values

```python
from sklearn.impute import SimpleImputer
```

```python
numeric_cols = spotify_yt_df_removed_column_NANvalues.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

```python
imputer = SimpleImputer(strategy = 'mean').fit(spotify_yt_df_removed_column_NANvalues[numeric_cols])
```

```python
spotify_yt_df_final=spotify_yt_df_removed_column_NANvalues.copy()
```

```python
spotify_yt_df_final[numeric_cols] = imputer.transform(spotify_yt_df_removed_column_NANvalues[numeric_cols])
```

```python
spotify_yt_df_final[numeric_cols].isna().sum()
```

#A Comparison between number of top view and likes for Artists

```python
labels = top_like['Artist']
view = top_like['Views']
like = top_like['Likes']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, view, width, label='view')
rects2 = ax.bar(x + width/2, like, width, label='like')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Artist')
ax.set_ylabel('Number of like and view for each Artist')
#ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

## Asking and Answering Questions





#### Q1: Which Artists are more popular in spotify and Youtube

For answering to this question, first, it is necessary to do groupby artists and select top streems for showig most popularity in spotify and top view showing most popularity in youtube.

```python
artist_views = spotify_yt_df_final.groupby('Artist')['Views'].sum()
artist_streams = spotify_yt_df_final.groupby('Artist')['Stream'].sum()
```

```python
top_artists_in_youtube = artist_views.sort_values(ascending=False).head(10)
top_artists_in_youtube
```

```python
top_artists_in_spotify = artist_streams.sort_values(ascending=False).head(10)
top_artists_in_spotify
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# top 10 yt
axes[0].set_title('Top 10 Artists on YouTube')
top_artists_in_youtube.plot(kind='bar', ax=axes[0])

# top 10 spotofy
axes[1].set_title('Top 10 Artists on Spotify')
top_artists_in_spotify.plot(kind='bar', ax=axes[1])


axes[0].set_xlabel('Artist')
axes[0].set_ylabel('Views')
axes[1].set_xlabel('Artist')
axes[1].set_ylabel('Streams')
fig.tight_layout()
plt.show()
```
